```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances, silhouette_score
from minisom import MiniSom
from sklearn.decomposition import PCA
```

```python

# Load the Iris dataset
iris = load_iris()
data = iris.data
target = iris.target

# Normalize the data
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data)

# Visualize the dataset using PCA (for 2D visualization)
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_normalized)

plt.figure(figsize=(10, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=target)
plt.title("PCA of Iris Dataset")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()

# Function to evaluate cluster homogeneity using mean intra-cluster distance
def evaluate_homogeneity(data_normalized, winner_coordinates, target):
    clusters = np.unique(winner_coordinates, axis=0)

    # Evaluate intra-cluster homogeneity
    for cluster in clusters:
        cluster_data = data_normalized[np.all(winner_coordinates == cluster, axis=1)]
        if cluster_data.shape[0] > 1:  # Ensure there is more than one data point in the cluster
            distances = pairwise_distances(cluster_data)
            mean_distance = np.mean(distances)
            print(f"Cluster {cluster} - Mean Intra-Cluster Distance: {mean_distance}")
        else:
            print(f"Cluster {cluster} - Mean Intra-Cluster Distance: N/A (single data point)")


# Function to evaluate outliers based on distance threshold
def evaluate_outliers(data_normalized, winner_coordinates, threshold=2.0):
    # Calculate distances from each data point to its BMU
    distances_to_bmu = np.array([np.linalg.norm(x - som.get_weights()[som.winner(x)]) for x in data_normalized])
    outliers = np.where(distances_to_bmu > threshold)[0]
    print(f"Outliers: {outliers}")
    plt.scatter(data_normalized[:, 0], data_normalized[:, 1], label='Normal Points')
    plt.scatter(data_normalized[outliers, 0], data_normalized[outliers, 1], color='red', label='Outliers')
    plt.legend()
    plt.show()


# Function to train SOM with different parameters using MiniSom
def train_som(data_normalized, grid_size=(10, 10), learning_rate=0.6, n_iter=100):
    som = MiniSom(x=grid_size[0], y=grid_size[1], input_len=data_normalized.shape[1], sigma=1.0, learning_rate=learning_rate)
    som.random_weights_init(data_normalized)
    som.train_random(data_normalized, num_iteration=n_iter)
    return som

# Function to calculate Silhouette Score for evaluating clustering quality
def evaluate_silhouette(data_normalized, som):
    # Get the BMU coordinates for each data point
    # Correctly access the grid size from the weights shape
    som_shape = som.get_weights().shape
    labels = [som.winner(x)[0] * som_shape[1] + som.winner(x)[1] for x in data_normalized]  # Cluster labels based on BMU

    # Calculate the Silhouette Score
    if len(np.unique(labels)) > 1:  # Silhouette score requires at least 2 clusters
        silhouette_avg = silhouette_score(data_normalized, labels)
        print(f"Silhouette Score: {silhouette_avg}")
        return silhouette_avg
    else:
        print("Silhouette Score: N/A (only one cluster found)")
        return None


# Test with different grid sizes, learning rates, and number of neurons
grid_sizes = [(10, 10), (15, 15), (20, 20)]  # Number of neurons: 100, 225, 400
learning_rates = [0.6, 0.4, 0.2]
n_iter = 100
silhouette_scores = []

for grid_size in grid_sizes:
    for lr in learning_rates:
        print(f"Training with grid size {grid_size} ({grid_size[0]*grid_size[1]} neurons) and learning rate {lr}")
        som = train_som(data_normalized, grid_size=grid_size, learning_rate=lr, n_iter=n_iter)

        # Visualize the results
        winner_coordinates = np.array([som.winner(x) for x in data_normalized])
        plt.figure(figsize=(10, 6))
        plt.scatter(winner_coordinates[:, 0], winner_coordinates[:, 1], c=target)
        plt.title(f"SOM with Grid Size {grid_size} and Learning Rate {lr}")
        plt.show()

        # Evaluate homogeneity of clusters for each setting
        evaluate_homogeneity(data_normalized, winner_coordinates, target)

        # Evaluate outliers based on distance threshold
        evaluate_outliers(data_normalized, winner_coordinates)

        # Evaluate Silhouette Score for clustering quality
        silhouette = evaluate_silhouette(data_normalized, som)
        if silhouette is not None:
            silhouette_scores.append((grid_size, lr, silhouette))

# Print the best silhouette score for model parameter tuning
if silhouette_scores:
    best_config = max(silhouette_scores, key=lambda x: x[2])
    print(f"Best Configuration - Grid Size: {best_config[0]}, Learning Rate: {best_config[1]}, Silhouette Score: {best_config[2]}")
else:
    print("No valid silhouette scores were calculated.")
```

```python
# Train SOM with the best parameters
best_grid_size = (10, 10)
best_learning_rate = 0.2
best_n_iter = 100 # Using the same number of iterations as before

som_best = MiniSom(x=best_grid_size[0], y=best_grid_size[1], input_len=data_normalized.shape[1], sigma=1.0, learning_rate=best_learning_rate)
som_best.random_weights_init(data_normalized)
som_best.train_random(data_normalized, num_iteration=best_n_iter)

print("SOM trained with best parameters.")
```

```python
# Visualize the U-Matrix of the trained SOM
plt.figure(figsize=(10, 10))
plt.pcolor(som_best.distance_map().T, cmap='bone_r')  # Transposed for better visualization
plt.colorbar()
plt.title("U-Matrix of the Trained SOM")
plt.show()
```

```python
# Visualize the trained SOM with data points
plt.figure(figsize=(10, 10))

# Get the winner coordinates for each data point
winner_coordinates = np.array([som_best.winner(x) for x in data_normalized])

# Plot the data points on the SOM grid, colored by their target labels
plt.scatter(winner_coordinates[:, 0], winner_coordinates[:, 1], c=target, cmap='viridis')

# Add a grid to the plot for better visualization of SOM nodes
plt.xticks(np.arange(best_grid_size[0] + 1) - 0.5, minor=True)
plt.yticks(np.arange(best_grid_size[1] + 1) - 0.5, minor=True)
plt.grid(which="minor", linestyle="-", color="black", linewidth=2)

plt.title("Data Points on Trained SOM")
plt.xlabel("SOM x-coordinate")
plt.ylabel("SOM y-coordinate")
plt.colorbar(label="Iris Species")
plt.show()
```

```python
# Visualize the trained SOM grid colored by the dominant class mapped to each neuron
plt.figure(figsize=(10, 10))

# Get the winning neuron for each data point
winner_coordinates = np.array([som_best.winner(x) for x in data_normalized])

# Create a dictionary to store the count of each class for each neuron
neuron_class_counts = {}
for i, coord in enumerate(winner_coordinates):
    coord_tuple = tuple(coord)  # Convert array to tuple for dictionary key
    if coord_tuple not in neuron_class_counts:
        neuron_class_counts[coord_tuple] = {0: 0, 1: 0, 2: 0} # Assuming 3 classes for Iris
    neuron_class_counts[coord_tuple][target[i]] += 1

# Create a grid to store the dominant class for each neuron
som_grid_classes = np.zeros((best_grid_size[0], best_grid_size[1]))

# Determine the dominant class for each neuron
for i in range(best_grid_size[0]):
    for j in range(best_grid_size[1]):
        coord_tuple = (i, j)
        if coord_tuple in neuron_class_counts:
            # Find the class with the maximum count for this neuron
            dominant_class = max(neuron_class_counts[coord_tuple], key=neuron_class_counts[coord_tuple].get)
            som_grid_classes[i, j] = dominant_class
        else:
            # If no data points mapped to this neuron, you might assign a default value or NaN
            som_grid_classes[i, j] = -1 # Use -1 to indicate no data points mapped

# Visualize the SOM grid colored by the dominant class
plt.pcolor(som_grid_classes.T, cmap='viridis', edgecolors='k', linewidths=0.5)
plt.colorbar(label="Iris Species (Dominant Class)")
plt.title("SOM Grid Colored by Dominant Iris Species")
plt.xticks(np.arange(best_grid_size[0]))
plt.yticks(np.arange(best_grid_size[1]))
plt.show()
```
